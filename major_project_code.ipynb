```python
# Define your variables
dependent_var = 'GDP Annual Growth'
independent_vars = [
    'Nifty 50 Annual Returns',
    'BSE Sensex Annual Returns',
    '10 - Year Bond Yield Returns',
    'Gold Annual Returns',
    'Inflation Annual'
]

# Set up the figure for subplots
plt.figure(figsize=(15, 10))
plt.suptitle('Simple Linear Regression: GDP Growth vs Investment Returns (2000-2023)', y=1.02, fontsize=14)

# Dictionary to store regression results
results = {}

for i, var in enumerate(independent_vars, 1):
    # Prepare data (drop missing values for each pair)
    temp_df = df[[dependent_var, var]].dropna()
    X = temp_df[var]
    y = temp_df[dependent_var]
    
    # Add constant for intercept term
    X = sm.add_constant(X)
    
    # Fit regression model
    model = sm.OLS(y, X).fit()
    results[var] = model
    
    # Create subplot
    plt.subplot(2, 3, i)
    sns.regplot(x=var, y=dependent_var, data=temp_df, 
                line_kws={'color': 'red'}, ci=None)
    
    # Add regression equation to plot
    intercept = model.params[0]
    slope = model.params[1]
    r_squared = model.rsquared
    equation = (f'y = {slope:.2f}x + {intercept:.2f}\n'
                f'R² = {r_squared:.2f}')
    
    plt.annotate(equation, xy=(0.05, 0.95), xycoords='axes fraction',
                ha='left', va='top', bbox=dict(boxstyle='round', fc='white'))
    
    plt.title(f'GDP Growth vs {var}')
    plt.xlabel(var)
    plt.ylabel(dependent_var)

plt.tight_layout()
plt.show()

# Print detailed regression results for each variable
print("\nRegression Results Summary:")
for var, model in results.items():
    print(f"\n=== {var} ===")
    print(model.summary())
```

```python
# Ensure data is sorted by year and set year as index
df = df.sort_values('Years').set_index('Years')

# Define maximum number of lags to test
max_lags = 4  # Typically 1-4 for annual data

# List of variables to test against GDP Growth
variables = [
    'Nifty 50 Annual Returns',
    'BSE Sensex Annual Returns', 
    '10 - Year Bond Yield Returns',
    'Gold Annual Returns',
    'Inflation Annual'
]

results = {}

for var in variables:
    print(f"\nGranger Causality: {var} → GDP Growth")
    print("="*50)
    
    # Create DataFrame with the two variables
    test_data = df[['GDP Annual Growth', var]].dropna()
    
    # Perform Granger test
    gc_result = grangercausalitytests(test_data, maxlag=max_lags, verbose=False)
    
    # Store results
    results[var] = gc_result
    
    # Print summary for each lag
    for lag in range(1, max_lags+1):
        p_value = gc_result[lag][0]['ssr_chi2test'][1]
        print(f"Lag {lag}: p-value = {p_value:.4f}", 
              "*" if p_value < 0.05 else "")
```

```python
# Prepare results for visualization
p_values = {var: [results[var][lag][0]['ssr_chi2test'][1] 
             for lag in range(1, max_lags+1)] 
             for var in variables}

# Create plot
plt.figure(figsize=(12, 6))
for var in variables:
    plt.plot(range(1, max_lags+1), p_values[var], 'o-', label=var)

plt.axhline(y=0.05, color='red', linestyle='--', alpha=0.5)
plt.xlabel('Number of Lags')
plt.ylabel('p-value')
plt.title('Granger Causality Test Results (GDP Growth)')
plt.legend()
plt.grid(True)
plt.show()
```
